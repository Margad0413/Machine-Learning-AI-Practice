* 시그모이드 함수: 이진 분류에 적합
* 소프트 맥스 함수: 다중 분류에 적합. 여러 함수의 출력값의 합이 1이 되도록 조정해서 확률 나오는거다. Z1 = z1/sum(z1:z7) 이런식임.
* 활성화 함수(시그모이드, 소프트맥스)는 로지스틱 회귀에서 알아서 정해준다.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

fish = pd.read_csv('http://bit.ly/fish_csv_data')
#print(fish) #pandas 하서 컬럼명이 존재

#print(fish['Species']) #Species 열의 값만 출력
#print(pd.unique(fish['Species'])) # 고유한 Species값들만 출력(pandas 제공)
#print(fish.head()) # 앞의 5값만 출력(pandas 제공)

fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()
fish_target = fish['Species'].to_numpy()

train_input, test_input, train_target, test_target = train_test_split(fish_input,fish_target, random_state=42)

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=20,max_iter=1000) #max_iter은 학습 반복 횟수. 100이 기본값.
#C는 규제제어. 작을수록 규제가 큰 것. 20에서 줄여나가자는 것.
lr.fit(train_scaled,train_target)

print(lr.score(train_scaled,train_target))
print(lr.score(test_scaled,test_target))

print(test_target[:5])
print(lr.predict(test_scaled[:5]))

print(lr.classes_)
print(np.round(lr.predict_proba(test_scaled[:5]),decimals=3))

print(lr.coef_.shape, lr.intercept_.shape) #7개의 특성이라서 총 7개의 함수가 있는거임.
#그 7개의 가중치와 절편으로 각각의 z값을 구하고, 시그모이드함수에 넣어서 0~1 사이의 값으로 변환후 값이 제일 높은걸로 예측된다.
#시그모이드로 2개의 값을 구해야하는 경우, 하나만 구하고 나머지는 1-상대확률 구하고 구하면 됨.
#그래서 시그모이드는 이진분류에 적합하고, 지금처럼 여러개일때(다중분류) 소프트맥스 함수 써야한다.
print(lr.coef_,lr.intercept_)
```
